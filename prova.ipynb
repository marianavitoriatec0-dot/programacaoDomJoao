```python
#Vamos fazer um programa para saber qual seria nosso peso em outro planeta.
print("Olá! Vamos ver qual seu peso em outro planeta/lua/corpo celeste?")
print("Para facilitar vamos utilizar numeros para identificar qual planeta, lua ou corpo celeste deseja, ok?")
print("Marte: 1")
print("Saturno: 2")
print("Vênus: 3")
print("Júpiter: 4")
print("Urano: 5")
print("Netuno: 6")
print("Mercúrio: 7")
print("Plutão : 8")
print("Lua: 9")

#Agora peço para o usuário escolher o planeta
planetaUsuario = int(input("Escolha um planeta(1 a 9): "))
if planetaUsuario == 1:
  print("Você escolheu o planeta Marte.")
elif planetaUsuario == 2:
  print("Você escolheu o planeta Saturno.")
elif planetaUsuario == 3:
  print("Você escolheu o planeta Vênus.")
elif planetaUsuario == 4:
  print("Você escolheu o planeta Júpiter.")
elif planetaUsuario == 5:
  print("Você escolheu o planeta Urano.")
elif planetaUsuario == 6:
  print("Você escolheu o planeta Netuno.")
elif planetaUsuario == 7:
  print("Você escolheu o planeta Mécurio")
elif planetaUsuario == 8:
  print("Você escolheu o corpo celeste plutão")
elif planetaUsuario == 9:
  print("Você escolheu a Lua.")



#Agora peço para o usuário colocar seu peso
pesoUsuario = float(input("Digite seu peso: "))

#Agora vou colocar em váriaveis os valores da gravidade dos planetas, pois precisamos dela para o calculo
marte = float(3.73)
saturno = float(10.44)
venus = float(8.87)
jupiter = float(24.79)
urano = float(8.69)
netuno = float(11.15)
mercurio = float(3.7)
plutao = float(0.62)
lua = float(1.62)

#Agora utilizaremos a gravidade e o peso para descobrimos

if planetaUsuario == 1:
  print(f"Seu peso em Marte é: {marte*pesoUsuario}kg ")
elif planetaUsuario == 2:
  print(f"Seu peso em Saturno é: {saturno*pesoUsuario}kg")
elif planetaUsuario == 3:
  print(f"Seu peso em Vênus é: {venus*pesoUsuario}kg")
elif planetaUsuario == 4:
  print(f"Seu peso em Júpiter é {jupiter*pesoUsuario}kg")
elif planetaUsuario == 5:
  print(f"Seu peso em Uruano é {urano*pesoUsuario}kg")
elif planetaUsuario == 6:
  print(f"Seu peso em Netuno é {netuno*pesoUsuario}kg ")
elif planetaUsuario == 7:
  print(f"Seu peso em Mécurio é {mercurio*pesoUsuario}kg")
elif planetaUsuario == 8:
  print(f"Seu peso em plutão é {plutao*pesoUsuario}kg ")
elif planetaUsuario == 9:
  print(f"Seu peso na Lua é {lua*pesoUsuario}kg")
```
